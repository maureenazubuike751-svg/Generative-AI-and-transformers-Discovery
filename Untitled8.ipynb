```python

# GENERATIVE AI & TRANSFORMERS - COMPLETE SUBMISSION

!pip install transformers torch

from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import torch

print("=" * 60)
print("GENERATIVE AI & TRANSFORMERS - COMPLETE SUBMISSION")
print("=" * 60)

# Sentiment Analysis
sentiment_pipeline = pipeline("sentiment-analysis")
print("✔ Sentiment pipeline ready")

# Text Generation
generation_pipeline = pipeline("text-generation", model="gpt2")
print("✔ Text generation pipeline ready")

# Summarization (custom to avoid errors)
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def summarize(text, max_length=50, min_length=20):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=1024)
    summary_ids = model.generate(inputs["input_ids"], max_length=max_length, min_length=min_length, do_sample=False)
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("✔ Summarization model ready")
print("=" * 60)

# Sentiment Analysis (6 sentences)
print("\nSENTIMENT ANALYSIS RESULTS:")
print("-" * 40)

test_sentences = [
    "I absolutely love this product!",
    "This is the worst experience ever.",
    "The movie was okay, nothing special.",
    "Oh great, another meeting.",
    "I'm so excited for the weekend!",
    "This traffic is making me furious."
]

for sentence in test_sentences:
    result = sentiment_pipeline(sentence)[0]
    print(f"Text: {sentence}")
    print(f"Sentiment: {result['label']} (confidence: {result['score']:.2f})")
    print("-" * 40)

# Text Generation
print("\nTEXT GENERATION:")
print("-" * 40)

prompt = "The future of artificial intelligence"
generated = generation_pipeline(prompt, max_length=50, num_return_sequences=1, do_sample=True, temperature=0.8)
print(generated[0]['generated_text'])
print("-" * 40)

# Summarization
print("\nSUMMARIZATION:")
print("-" * 40)

article = """
Hospitals are using AI to analyze medical images, predict patient outcomes,
and streamline administrative tasks. Recent studies show that AI algorithms
can detect certain types of cancer with accuracy comparable to human radiologists
in some cases. However, experts caution that AI should be used to assist,
not replace, healthcare professionals. Integrating AI into healthcare systems
requires careful planning, robust data privacy concerns, and the need for
continuous validation. Despite these challenges, the potential benefits include
faster diagnoses, personalized treatment plans, and reduced healthcare costs.
"""

print("Original Article:")
print(article)
print("-" * 40)

print("\nSummary (default - 50 tokens):")
print(summarize(article, max_length=50, min_length=20))

print("\nSummary (short - 30 tokens):")
print(summarize(article, max_length=30, min_length=10))

print("\nSummary (long - 100 tokens):")
print(summarize(article, max_length=100, min_length=50))

print("\n" + "=" * 60)
print("PROJECT COMPLETE - ALL REQUIREMENTS MET")
print("=" * 60)
```
